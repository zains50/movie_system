```python
import json 
import psycopg2

## DATA BASE CONNECTION
with open ("data/db_config.json") as f:
    db_config = json.load(f)

conn = psycopg2.connect(**db_config)
conn.autocommit = True 
cursor = conn.cursor()
```

```python
def execute_sql_from_file(file_name):
    with open(f"sql_scripts/{file_name}") as f:
        cursor.execute(f.read())
```

```python
execute_sql_from_file("get_movie_genres.sql")
cursor.fetchall()
```
